```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://hubertiming.com/results/2018MLK'

html = urlopen(url)
```

```python
soup = BeautifulSoup(html)
soup
```

```python
websiteTitle = soup.title

print(f"title element : {websiteTitle}\ncontent inside title : {websiteTitle.text}")
```

```python
allLinks = soup.find_all('a', href=True)

print(f"All the links on the page : \n {allLinks}\n\nDatatype of output : {type(allLinks)} (yeah, but basically a list!)")
```

```python
for link in allLinks:
    print(link['href']) # or link.get('href')
```

```python
allRows = soup.find_all("tr")

#! Participants starting from index 5

allRows[5].prettify()
```

```python
allRows[5]
```

```python
firstContestantRow = allRows[13]

row_list = firstContestantRow.find_all("td")

contestantData = []

for i in range(13):
    contestantData.append(" ".join(row_list[i].text.split()))

print(row_list[9].text.split(" "))

print(contestantData)
```

```python
randomContestantRow = allRows[np.random.randint(5, 150)]

row_list = randomContestantRow.find_all("td")

" ".join(row_list[2].text.split())
row_list[9] = row_list[9].text.split(" ")[1]
row_list[9]
```

```python
contestantData = []

for cell in allRows[5:]:
    rowEl = cell.find_all("td")
    Data = []

    # rowEl[9] = 

    for i in range(13):
        Data.append(" ".join(rowEl[i].text.split()))
    
    if rowEl[9].text != '':
        Data[9] = rowEl[9].text.split(" ")[1]
    contestantData.append(Data)

for candidate in contestantData:
    print(candidate, end='\n')
```

```python
contestant_df = pd.DataFrame(contestantData)

contestant_df
```

```python
contestant_df = pd.DataFrame(contestantData)
# contestant_df.columns['Place', 'Bib ID', 'Name', 'Gender', 'Age', 'City', 'Chip Time', 'Chip Pace', 'Gender Rank', 'Age Group', 'Age Group Rank', 'Time to Start', 'Gun Time']

contestant_df.to_csv('RaceData.csv',index=False)
```
