```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://hubertiming.com/results/2018MLK'

html = urlopen(url)
```

```python
soup = BeautifulSoup(html)
soup
```

```python
titleWebsite = soup.title

print(f"title element : {titleWebsite}\ncontent inside title : {titleWebsite.text}")
```

```python
allLinks = soup.find_all('a')

print(f"All the links on the page : \n {allLinks}\n\nDatatype of output : {type(allLinks)} (yeah, but basically a list!)")
```

```python
for link in allLinks:
    print(link['href'])
```
