```python
xC = np.array([1,2,3,4])
yC = np.array([5,6,7,8])

plt.plot(xC,yC,color='lime')
plt.show()
```

```python
points = np.array([4,1,5,2,3])

plt.plot(points, marker='o', color="lime")
plt.show()
```

```python
points = np.array([4,1,5,2,3])

plt.plot(points, marker='o', color="magenta",linestyle='dashdot')
plt.show()
```

```python
points = np.array([4,1,5,2,3])

plt.plot(points, marker='x', color="lime",linewidth='1.5',linestyle="dashdot")
plt.show()
```

```python
x1 = np.array([1,2,3,4])
x2 = np.array([3,6,8,10])
x3 = np.array([2,3,5,9])

y1 = np.array([1,4,6,8])
y2 = np.array([1,2,3,4])
y3 = np.array([1,5,7,11])

plt.plot(x1,y1,x2,y2,x3,y3)
plt.show()
```

```python
x = np.array([80,90,100,110])
y = np.array([200,230,260,270])

plt.plot(x,y, color='lime')
plt.title("Sports data")
plt.xlabel("Sports")
plt.ylabel("Data")

plt.show()
```

```python
x = np.array([80,90,100,110])
y = np.array([200,230,260,270])

plt.plot(x,y, color='lime')
plt.title("Sports data")
plt.xlabel("Sports")
plt.ylabel("Data")

plt.grid(axis='y')
plt.show()
```
