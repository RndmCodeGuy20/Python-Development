```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
colors = ['#003f5c', '#2f4b7c', '#665191', '#a05195', '#d45087', '#f95d6a', '#ff7c43', '#ffa600']
```

```python
sales_df = pd.read_csv('../Assets/Company-Sales.csv')
sales_df
```

```python
months_ls = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(20,5))
plt.plot(months_ls, sales_df['total_profit'], marker='o', color=colors[0], ls='-.', mfc=colors[7], alpha=0.8)

plt.xlabel('Months in Year')
plt.ylabel('Total Profit')

plt.show()
```

```python
months_ls = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(20,5))
plt.plot(months_ls, sales_df['total_profit'], marker='o', color='red', ls=':', mfc=colors[3], alpha=0.8, lw=3, )

plt.xlabel('Months in Year')
plt.ylabel('Sold Units Number')
plt.legend(['Monthly Profits'], loc=4)

plt.show()
```

```python
sns.set_style('darkgrid')
```

```python
months_ls = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(20,7.5))
plt.plot(months_ls, sales_df['facecream'], marker='o', color=colors[0], ls='-.', mfc=colors[7], alpha=0.8)
plt.plot(months_ls, sales_df['facewash'], marker='o', color=colors[1], ls='-.', mfc=colors[6], alpha=0.8)
plt.plot(months_ls, sales_df['toothpaste'], marker='o', color=colors[2], ls='-.', mfc=colors[5], alpha=0.8)
plt.plot(months_ls, sales_df['bathingsoap'], marker='o', color=colors[3], ls='-.', mfc=colors[4], alpha=0.8)
plt.plot(months_ls, sales_df['shampoo'], marker='o', color=colors[4], ls='-.', mfc=colors[3], alpha=0.8)
plt.plot(months_ls, sales_df['moisturizer'], marker='o', color=colors[5], ls='-.', mfc=colors[2], alpha=0.8)

plt.xlabel('Months in Year')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold Per Month')
plt.legend(['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.show()
```

```python
months_ls = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(10,10))

sales_data = [sales_df['facecream'].sum(), sales_df['facewash'].sum(), sales_df['toothpaste'].sum(), sales_df['bathingsoap'].sum(), sales_df['shampoo'].sum(), sales_df['moisturizer'].sum()]

plt.axis('equal')
plt.pie(sales_data, colors=colors)

plt.legend(['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.title('Sales Data of the Year')

plt.show()
```
